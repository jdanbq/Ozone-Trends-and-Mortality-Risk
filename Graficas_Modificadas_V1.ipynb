```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
import matplotlib.gridspec as gridspec
```

```python
# Lista de nombres de estaciones

estaciones = [ 'CARVAJAL','CIUDAD BOLIVAR', 'FONTIBON', 'GUAYMARAL','LAS FERIAS','MIN AMBIENTE','SAN CRISTOBAL','SUBA','USAQUEN']
# Nombres de las variables
estacionesB= ['CENTRO DE ALTO RENDIMIENTO', 'JAZMIN', 'TUNAL', 'KENNEDY', 'USME','PUENTE ARANDA']
```

```python
directory = 'Predicciones/'  # Ruta al directorio que contiene los archivos CSV

for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        file_path = os.path.join(directory, filename)

        try:
            # Lee el archivo CSV con el punto como separador decimal y la coma como separador de columnas
            df = pd.read_csv(file_path, sep=',', decimal='.')
            print(f'Contenido de {filename}:')
            print(df)
        except FileNotFoundError:
            print(f'El archivo "{file_path}" no se encontró.')
        except Exception as e:
            print(f'Ocurrió un error al intentar abrir el archivo {filename}: {str(e)}')
```

```python
num_filas = 3
num_columnas = 3

fig = plt.figure(figsize=(12, 10))
gs = gridspec.GridSpec(num_filas, num_columnas, figure=fig, 
                       left=0.05, right=0.975, hspace=0.25, wspace=0.2, top=0.9, bottom=0.1)

# Generar fechas para los últimos 100 días de 2022
dates = pd.date_range(start='2022-09-23', periods=100, freq='D')

for i, estacion in enumerate(estaciones):
    file_path = os.path.join(directory, f'prediccion_estacion_{estacion}.csv')

    try:
        df = pd.read_csv(file_path, sep=',', decimal='.')

        fila = i // num_columnas
        columna = i % num_columnas

        ax = fig.add_subplot(gs[fila, columna])

        ax.set_title(f'{chr(97 + i)}) {estacion}', fontsize=14)
        ax.plot(dates, df['Prediccion'][0:100], label='Ozone prediction', color='purple')
        ax.plot(dates, df['Real'][0:100], label='Ozone observed', color='tomato')

        ax.set_yticks(np.arange(8, 23, 2))

        if columna == 0:
            ax.set_ylabel('μg m$^{-3}$', fontsize=12)
            ax.tick_params(axis='y', labelsize = 12)
        else:
            ax.set_yticks([])

        if fila == num_filas - 1:
            ax.xaxis.set_major_locator(plt.MaxNLocator(6))
            ax.tick_params(axis='x', rotation=90, labelsize = 12)
        else:
            ax.xaxis.set_major_locator(plt.MaxNLocator(6))
            ax.tick_params(axis='x', rotation=90)
            ax.set_xticklabels([])

        if fila == 0:
            ax.legend(frameon=False, bbox_to_anchor=(0.5, 1.32), loc='upper center', ncol=4, fontsize=12)
        if columna == 0 or columna == 2:
            ax.legend('', frameon = False)

    except FileNotFoundError:
        print(f'El archivo "{file_path}" no se encontró.')
    except Exception as e:
        print(f'Ocurrió un error al intentar abrir el archivo {file_path}: {str(e)}')

# Eliminar subplots vacíos si hay menos estaciones que espacio en la cuadrícula
total_subplots = num_filas * num_columnas
for i in range(len(estaciones), total_subplots):
    fila = i // num_columnas
    columna = i % num_columnas
    ax = fig.add_subplot(gs[fila, columna])
    fig.delaxes(ax)

plt.show()
```

```python
directory = 'Predicciones/'
num_filas = 3  # según tu for-loop (0 a 3)
num_columnas = 3

fig, axs = plt.subplots(num_filas, num_columnas, figsize=(10, 12), sharex=True, sharey=True)

for i in range(num_filas):
    for j in range(num_columnas):
        ax = axs[i, j]
        index = i * num_columnas + j
        if index >= len(estaciones):
            ax.axis('off')
            continue

        estacion = estaciones[index]

        try:
            file_path = os.path.join(directory, f'prediccion_estacion_{estacion}.csv')
            df = pd.read_csv(file_path, sep=',', decimal='.')

            real = df['Real'][0:100]
            prediccion = df['Prediccion'][0:100]

            ax.scatter(real, prediccion, color='orange', s=5)
            ax.set_title(f'{chr(97 + index)}) {estacion}', loc='left', fontsize=16)

            desviacion_estandar = np.std(real)

            ymax = max(max(real), max(prediccion))
            ax.plot([0, ymax], [0, ymax], color="black", lw=3, label='Línea de Desviación')
            ax.plot([0, ymax], [0, ymax - desviacion_estandar / 2], color="lightcoral", linestyle='--', label='Línea Inferior')
            ax.plot([0, ymax], [0, ymax + desviacion_estandar * 2], color="cornflowerblue", linestyle='--', label='Línea Superior')

            ax.set_xlim(0, ymax)
            ax.set_ylim(0, ymax)
            ax.set_xticks(np.arange(0, ymax+1, 5))
            ax.set_yticks(np.arange(0, ymax+1, 5))
            ax.tick_params(axis='both', labelsize=12)

            # Mostrar etiquetas X sólo en la última fila
            if i == num_filas - 1:
                ax.set_xlabel('Observations', fontsize=15)
            else:
                ax.set_xlabel('')
                ax.tick_params(axis='x', labelbottom=False)

            # Mostrar etiquetas Y sólo en la primera columna
            if j == 0:
                ax.set_ylabel('Prediction', fontsize=15)
            else:
                ax.set_ylabel('')
                ax.tick_params(axis='y', labelleft=False)

        except FileNotFoundError:
            ax.axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Directorio con los archivos CSV
directory = 'Predicciones/'

# Configura el número de filas y columnas para organizar los subplots
num_filas = 4
num_columnas = 3

fig, axs = plt.subplots(num_filas, num_columnas, figsize=(15, 20))
fig.suptitle("Scatter Plots de Estaciones", fontsize=16)

for i in range(4):
    for j in range(3):
        ax = axs[i, j]
        estacion = estaciones[i * 3 + j]  # Calcula la estación actual

        try:
            file_path = os.path.join(directory, f'prediccion_estacion_{estacion}.csv')
            df = pd.read_csv(file_path, sep=',', decimal='.')

            # Calcula los valores para el scatter plot
            real = df['Real'][0:100]
            prediccion = df['Prediccion'][0:100]

            ax.scatter(real, prediccion, color='orange', s=5)
            ax.set_title(f'{chr(97 + i * 3 + j)}) {estacion}', loc='left', fontsize=18)
            ax.set_xlabel('Valores Reales', fontsize=15)
            ax.set_ylabel('Valores Predichos', fontsize=15)
            
            if i == 3:
                ax.set_xticks(np.arange(0, 101, 20))  # Establece los ticks y etiquetas en el eje X
                ax.set_xticklabels(ax.get_xticks(), fontsize=12, fontweight='bold')
            else:
                ax.set_xticks([])  # Elimina los ticks del eje X para las filas superiores de subplots

            if j == 0:
                ax.set_yticks(np.arange(0, 101, 20))  # Establece los ticks y etiquetas en el eje Y
                ax.set_yticklabels(ax.get_yticks(), fontsize=12, fontweight='bold')
            else:
                ax.set_yticks([])  # Elimina los ticks del eje Y para las columnas derechas de subplots
        except FileNotFoundError:
            ax.axis('off')  # Si no se encuentra el archivo, muestra una subtrama vacía

plt.tight_layout()
plt.show()
```
